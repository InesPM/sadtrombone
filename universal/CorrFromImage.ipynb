Can you get autocorrelations from pngs of FRBs?
===

yes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
import os
from tqdm import tqdm
import corrfns

import matplotlib.image as mpimg

# imfrb = imageio.imread('frb180814_oct28.png')
imfrb = mpimg.imread('frb180814_oct28.png')
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
grayfrb = 1 - rgb2gray(imfrb) # 1 is bright, 0 is dim
```

```python
# burstwindow = grayfrb[451:] - grayfrb[451:, 0:50].mean(axis=1)[:,None]
burstwindow = grayfrb - grayfrb[:, 0:50].mean(axis=1)[:,None]
plt.figure(figsize=(4, 8))
plt.imshow(burstwindow, aspect='equal', cmap='gray_r') # white is 0, black is 1
plt.colorbar()
# plt.axhline(y=85)
# plt.axhline(y=95)
# grayfrb[85:95]
corr = corrfns.auto_corr2D_viafft(burstwindow)
plt.figure(figsize=(8, 8))
plt.imshow(corr, aspect='equal', cmap='gray')
plt.colorbar()
```
